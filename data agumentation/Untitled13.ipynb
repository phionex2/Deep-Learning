```python
from tensorflow import keras
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('OIP.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
type(img)
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
import os
os.mkdir("augu")
```

```python
i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='augu'):
  i = i + 1
  if i == 10:
    break
```
